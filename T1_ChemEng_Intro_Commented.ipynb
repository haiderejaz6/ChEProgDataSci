## Problem 1 — Reactor Temperature Readings

A reactor recorded temperatures (K) each minute for 8 minutes:

`T = [350, 352, 351, 355, 358, 360, 359, 361]`

**Tasks**
1. Store the data in a list.
2. Compute the average, min, and max.
3. Check if the reactor is warming overall.


```python

# Store the recorded reactor temperatures in a list
T = [350, 352, 351, 355, 358, 360, 359, 361]

# Calculate average temperature (sum divided by number of readings)
avg_T = sum(T) / len(T)

# Find minimum and maximum temperatures in the list
min_T = min(T)
max_T = max(T)

print("Average T (K):", avg_T)
print("Minimum T (K):", min_T)
print("Maximum T (K):", max_T)

# Check if reactor is warming overall: compare last and first reading
warming = T[-1] > T[0]
print("Warming overall?", warming)
```


### Optional Plot — Temperature vs. Time


```python

import matplotlib.pyplot as plt

# Create time steps corresponding to measurements (0 to 7 minutes)
time_min = list(range(len(T)))

# Plot temperature vs. time
plt.plot(time_min, T, marker='o')
plt.xlabel("Time (min)")
plt.ylabel("Temperature (K)")
plt.title("Reactor Temperature vs. Time")
plt.show()
```


## Problem 2 — Density Calculation

Density = m / V

Suppose m = 500 kg and V = 0.65 m³.


```python

# Define mass in kilograms
m = 500.0   # kg

# Define volume in cubic meters
V = 0.65    # m^3

# Calculate density (rho = m/V)
rho = m / V

print("Density (kg/m³):", rho)
```


## Problem 3 — Temperature Conversions

Convert 80 °C to K and °F.


```python

# Define temperature in Celsius
T_C = 80.0

# Convert Celsius to Kelvin
T_K = T_C + 273.15

# Convert Celsius to Fahrenheit
T_F = T_C * 9/5 + 32

print("Temperature:", T_C, "°C")
print("In Kelvin:", T_K, "K")
print("In Fahrenheit:", T_F, "°F")
```


## Problem 4 — Ideal Gas Law

PV = nRT

Given:
- P = 5 bar
- V = 2 m³ = 2000 L
- T = 300 K
- R = 0.08314 bar·L/(mol·K)

Compute moles (n).


```python

# Define pressure in bar
P_bar = 5.0

# Define volume in cubic meters
V_m3 = 2.0

# Define temperature in Kelvin
T_K = 300.0

# Define gas constant in bar·L/(mol·K)
R_barL = 0.08314

# Convert volume from m³ to liters
V_L = V_m3 * 1000.0

# Calculate number of moles (n = PV/RT)
n = (P_bar * V_L) / (R_barL * T_K)

print("Volume (L):", V_L)
print("Moles of gas:", n)
```


## Mini-Challenge — Mole Fractions

Given:
- Mass_A = 36 g, MW_A = 18 g/mol
- Mass_B = 92 g, MW_B = 46 g/mol

Compute mole fractions.


```python

# Define a function to compute mole fractions for a binary mixture
def mole_fractions(mass_A, mass_B, MW_A, MW_B):
    # Convert masses to moles
    n_A = mass_A / MW_A
    n_B = mass_B / MW_B

    # Calculate total moles
    total = n_A + n_B

    # Mole fractions of A and B
    x_A = n_A / total
    x_B = n_B / total

    return x_A, x_B

# Test the function with given values
xA, xB = mole_fractions(36, 92, 18, 46)

print("x_A:", xA, "x_B:", xB)
```
