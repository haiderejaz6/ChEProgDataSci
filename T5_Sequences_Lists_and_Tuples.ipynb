## 5.2 Lists
A **list** is a collection of ordered items that can be changed (mutable).  
You can store numbers, strings, or even different types in one list.


````python
# Example: a list of integers
c = [-45, 6, 0, 72, 1543]
print(c)

# Example: a list with mixed data types
student = ['Mary', 'Smith', 3.57, 2022]
print(student)
````

### Accessing Elements
You can access elements in a list using square brackets `[]`.  
Indexing starts from **0** (the first element).  
For example, `c[0]` gets the first element, and `c[4]` gets the fifth.


````python
print(c[0])  # first element
print(c[4])  # fifth (last) element
````

### Negative Indices
You can also count from the end using **negative indices**.  
`c[-1]` means the last element, `c[-2]` means the second last, and so on.


````python
print(c[-1])  # last element
print(c[-5])  # first element (since there are 5 total)
````

### Determining Length
Use the `len()` function to find how many elements a list contains.


````python
len(c)
````

### Index Rules
Indices must be integers or integer expressions.  
You can calculate an index using expressions like `a + b`.


````python
a, b = 1, 2
print(c[a + b])  # equivalent to c[3]
````

### Mutability
Lists are **mutable**, meaning you can change their elements after creation.


````python
print("Before:", c)
c[4] = 17
print("After:", c)
````

### Immutable Sequences
Strings and tuples are **immutable**, you can access their elements, but you **cannot change** them.


````python
s = 'hello'
print(s[0])   # 'h'

# Trying to change a character will cause an error
# Uncomment to see the error:
# s[0] = 'H'
````

### Index Errors
If you try to access a position that doesn’t exist, Python raises an `IndexError`.


````python
# Uncomment the next line to see the error
# c[100]
````

### Using Elements in Expressions
List elements can be used in expressions just like variables.


````python
result = c[0] + c[1] + c[2]
print(result)
````

### Appending with `+=`
You can use the `+=` operator to add elements to a list.  
The list grows dynamically as you add items.


````python
a_list = []
for number in range(1, 6):
    a_list += [number]
print(a_list)
````

### Appending Strings and Tuples
When the right-hand side of `+=` is an iterable (like a string or tuple),  
each element is added individually to the list.


````python
letters = []
letters += 'Python'
print(letters)

numbers = []
numbers += (10, 20, 30)
print(numbers)
````

### Concatenating Lists
Use the `+` operator to join two lists into a new one.  
The original lists remain unchanged.


````python
list1 = [10, 20, 30]
list2 = [40, 50]
concatenated = list1 + list2
print(concatenated)
````

### Looping with Indices
Use `range(len(list))` to loop through list indices and values.  
This is useful when you need both the **position** and the **element**.


````python
for i in range(len(concatenated)):
    print(f'Index {i}: {concatenated[i]}')
````

### Comparing Lists
Lists can be compared using comparison operators (`==`, `<`, `>`, etc.).  
Python compares them **element by element** from left to right.


````python
a = [1, 2, 3]
b = [1, 2, 3]
c = [1, 2, 3, 4]

print(a == b)  # True: same elements
print(a == c)  # False: c has an extra element
print(a < c)   # True: a is shorter
print(c >= b)  # True: first 3 elements equal, c is longer
````

## 5.3 Tuples
A **tuple** is an immutable sequence used to store multiple items.  
Tuples are often used for grouping related pieces of data, like a student’s name and GPA.  
Unlike lists, tuples **cannot be changed** after creation (no adding, removing, or modifying elements).


````python
# Creating an empty tuple
student_tuple = ()
print(student_tuple)
print("Length:", len(student_tuple))
````

## Tuples vs Lists

Both **tuples** and **lists** are sequence types that can store ordered collections of items —  
but they differ in several important ways.


````python
# A list (uses square brackets)
num_list = [10, 20, 30]

# A tuple (uses parentheses)
num_tuple = (10, 20, 30)

print("List:", num_list)
print("Tuple:", num_tuple)

print(type(num_list))
print(type(num_tuple))
````

### Creating Tuples with Values
Tuples can be created by separating values with commas — parentheses are optional but recommended for clarity.


````python
student_tuple = 'John', 'Green', 3.3
print(student_tuple)
print("Length:", len(student_tuple))

another_student_tuple = ('Mary', 'Red', 3.3)
print(another_student_tuple)
````

### One-Element Tuples
To create a **one-element tuple**, you must include a comma after the single value.  
Without the comma, Python treats it as a simple value, not a tuple.


````python
a_singleton_tuple = ('red',)
print(a_singleton_tuple)

not_a_tuple = ('red')
print(type(not_a_tuple))  # just a string, not a tuple
````

### Accessing Tuple Elements
Tuples use zero-based indexing just like lists.  
You can access individual elements with `[]` but you **cannot modify** them.


````python
time_tuple = (9, 16, 1)
print(time_tuple)

# Calculate total seconds since midnight
seconds_since_midnight = time_tuple[0] * 3600 + time_tuple[1] * 60 + time_tuple[2]
print(seconds_since_midnight)
````

### Tuple Immutability
Trying to assign a new value to a tuple element raises a `TypeError`.


````python
# Uncomment the line below to see the error
# time_tuple[0] = 10
````

### Concatenation with `+=`
Tuples can be concatenated using `+=`, which creates a **new tuple**.  
The original tuple remains unchanged.


````python
tuple1 = (10, 20, 30)
tuple2 = tuple1  # both variables refer to the same object

print("Before concatenation:")
print("tuple1:", tuple1)
print("tuple2:", tuple2)
print("tuple1 is tuple2:", tuple1 is tuple2)  # True — same object in memory

# Now concatenate
tuple1 += (40, 50)

print("\nAfter concatenation:")
print("tuple1:", tuple1)
print("tuple2:", tuple2)
print("tuple1 is tuple2:", tuple1 is tuple2)  # False — tuple1 now points to a *new* object
````

### Type Matching with `+=`
When using `+=` with tuples or strings, the right-hand side must be of the same type.  
Mixing a tuple with a list or string causes a `TypeError`.


````python
# Uncomment to see the TypeError
# tuple1 += [60, 70]   # can't add list to tuple
````

### Appending Tuples to Lists
You can use `+=` to append a tuple’s elements to a **list** — the list expands dynamically.


````python
numbers = [1, 2, 3, 4, 5]
numbers += (6, 7)
print(numbers)
````

### Tuples Can Contain Mutable Objects
Even though tuples themselves are immutable, they can hold **mutable objects** (like lists).  
You can modify the mutable element inside the tuple.


````python
student_tuple = ('Amanda', 'Blue', [98, 75, 87])
print(student_tuple)

# Modify the list inside the tuple
student_tuple[2][1] = 85
print(student_tuple)
````

### How Nested Access Works
In `student_tuple[2][1]`,  
- `student_tuple[2]` accesses the list `[98, 75, 87]`,  
- `[1]` accesses its second element (75),  
- and the assignment changes it to 85.


````python
# Example to demonstrate nested access
print("Inner list before:", student_tuple[2])
student_tuple[2][1] = 90
print("Inner list after :", student_tuple[2])
````

## 5.4 Unpacking Sequences
**Unpacking** allows you to assign elements of any sequence (list, tuple, string, or range)  
directly to multiple variables in a single statement.  
The number of variables on the left must exactly match the number of elements in the sequence.


````python
student_tuple = ('Amanda', [98, 85, 87])

first_name, grades = student_tuple
print(first_name)
print(grades)
````

### Unpacking Different Sequences
You can unpack elements from any sequence - strings, lists, tuples, or ranges.


````python
# Unpacking a string
first, second = 'hi'
print(first, second)

# Unpacking a list
number1, number2, number3 = [2, 3, 5]
print(number1, number2, number3)

# Unpacking a range
number1, number2, number3 = range(10, 40, 10)
print(number1, number2, number3)
````

### ValueError: Mismatch in Number of Variables
If the number of variables and sequence elements don’t match,  
Python raises a `ValueError`.


````python
# Uncomment to see the error
# a, b = (1, 2, 3)  # too many values to unpack
````

### Swapping Values via Packing and Unpacking
You can **swap variable values** easily without a temporary variable  
by packing them into a tuple on the right and unpacking on the left.


````python
number1 = 99
number2 = 22

print(f'number1 = {number1}; number2 = {number2}')

number1, number2 = (number2, number1)
print(f'number1 = {number1}; number2 = {number2}')
````

## Accessing Indices and Values Safely with `enumerate()`
The built-in function **`enumerate()`** allows you to loop over a sequence  
while keeping track of both the **index** and the **value**.  
It prevents errors from using incorrect index ranges with `range()` and `len()`.


````python
colors = ['red', 'orange', 'yellow']

# Create a list of (index, value) pairs
print(list(enumerate(colors)))

# Create a tuple of (index, value) pairs
print(tuple(enumerate(colors)))
````

### Looping with `enumerate()`
When looping, you can unpack each `(index, value)` tuple returned by `enumerate()`.  
This is safer and more readable than manually managing indices.


````python
for index, value in enumerate(colors):
    print(f'{index}: {value}')
````

## Creating a Primitive Bar Chart
We can use `enumerate()` to display both the index and value of list elements,  
and represent the value visually using asterisks (`*`).  
The expression `"*" * value` repeats the string `"*"` a number of times equal to the element value.


````python
# fig05_01.py — Creating a simple text-based bar chart

# A list of numbers representing the values to be visualized as bars
numbers = [19, 3, 15, 7, 11]

# Print a blank line (\n), then a title for the bar chart
print('\nCreating a bar chart from numbers:')

# Print the column headers:
# "Index" and "Value" are aligned with formatted spacing.
# The expression {"Value":>8} means "right-align the word 'Value' in 8 spaces."
print(f'Index{"Value":>8}   Bar')

# Use enumerate() to loop through both the index and value of each element in the list
for index, value in enumerate(numbers):

    # For each number, print:
    # 1. The index (right-aligned in 5 spaces: {index:>5})
    # 2. The value (right-aligned in 8 spaces: {value:>8})
    # 3. A bar made of asterisks '*' repeated 'value' times
    #    ("*" * value creates a string like "***" or "*******")
    print(f'{index:>5}{value:>8}   {"*" * value}')
````

## 5.5 Sequence Slicing
**Slicing** allows you to extract portions (subsequences) of sequences like lists, tuples, or strings.  
A slice creates a **new sequence** of the same type containing the selected elements.  
Mutable sequences (like lists) can also be **modified** using slices.


````python
numbers = [2, 3, 5, 7, 11, 13, 17, 19]

# Basic slice — elements at indices 2 through 5 (6 is excluded)
print(numbers[2:6])  
````

### Slice Syntax
General form: `sequence[start:end:step]`

- **start** → index where the slice begins (inclusive)  
- **end** → index where the slice ends (exclusive)  
- **step** → how many items to skip (optional)


````python
numbers = [2, 3, 5, 7, 11, 13, 17, 19]

# Example: indices 2 to 6 (end is exclusive)
print(numbers[2:6])
````

### Omitting the Starting Index
If the starting index is omitted, Python assumes it’s `0`.


````python
numbers = [2, 3, 5, 7, 11, 13, 17, 19]

print(numbers[:6])   # same as numbers[0:6]
````

### Omitting the Ending Index
If the ending index is omitted, Python assumes it’s the sequence’s length.


````python
numbers = [2, 3, 5, 7, 11, 13, 17, 19]

print(numbers[6:])         # from index 6 to the end
print(numbers[6:len(numbers)])  # same result
````

### Omitting Both Indices
If both start and end are omitted, the slice copies the **entire sequence**.


````python
numbers = [2, 3, 5, 7, 11, 13, 17, 19]

copy_numbers = numbers[:]
print(copy_numbers)
````

Note: This creates a **shallow copy** — the new list references the same inner objects,  
not duplicates of them.


### Using a Step
The third parameter, `step`, lets you skip elements while slicing.


### Negative Indices and Steps
Negative indices count from the end.  
A **negative step** reverses the order of elements.


````python
# Reversing a list using slicing
print(numbers[::-1])        # [19, 17, 13, 11, 7, 5, 3, 2]
print(numbers[-1:-9:-1])    # same result using explicit indices
````

## Modifying Lists with Slices
Slicing can also **replace, delete, or insert** elements in mutable sequences like lists.


````python
numbers = [2, 3, 5, 7, 11, 13, 17, 19]

# Replace first three elements
numbers[0:3] = ['two', 'three', 'five']
print(numbers)
````

### Deleting Elements Using an Empty List
Assigning an empty list `[]` to a slice deletes that portion.


````python
numbers = [2, 3, 5, 7, 11, 13, 17, 19]

numbers[0:3] = []
print(numbers)
````

### Assigning to a Stepped Slice
You can modify multiple spaced elements at once using a step.


````python
numbers = [2, 3, 5, 7, 11, 13, 17, 19]
numbers[::2] = [100, 100, 100, 100]  # replaces every 2nd element
print(numbers)
````

### Deleting All Elements with a Slice
Assigning `[]` to a full slice clears the list but keeps the same object in memory.


````python
numbers = [2, 3, 5, 7, 11, 13, 17, 19]

print("Before:", id(numbers))
numbers[:] = []
print("After clearing:", numbers)
print("Same object ID:", id(numbers))
````

### Assigning a New Empty List
Assigning `numbers = []` creates a **new empty list object**,  
so its memory ID changes.


````python
numbers = [2, 3, 5, 7, 11, 13, 17, 19]

numbers = []
print(numbers)
print("New object ID:", id(numbers))
````

## 5.6 The `del` Statement
The `del` statement is used to **delete elements, slices, or entire variables**.  
It can remove items from a list by index or slice, or completely delete a variable from memory.


````python
# Create a list of numbers from 0 to 9
numbers = list(range(0, 10))
print(numbers)
````

### Deleting a Slice from a List

### Deleting with a Step

### Deleting All Elements (Clearing the List)

This empties the list but keeps the list object itself.


````python
del numbers[:]
print(numbers)
````

### Deleting a Variable

````python
del numbers

# Uncomment the next line to see the error
# print(numbers)  # NameError: name 'numbers' is not defined
````

## 5.7 Passing Lists to Functions
In Python, **everything is passed by reference** , meaning that function arguments refer to the same object in memory.  
If the object is **mutable** (like a list), changes made inside the function affect the original object.


````python
# Define a function that multiplies each element in a list by 2
def modify_elements(items):
    """Multiplies all element values in items by 2."""
    for i in range(len(items)):
        items[i] *= 2
````

### Passing an Entire List
Here, we pass a list to `modify_elements()`.  
Because lists are **mutable**, the function modifies the original list directly.


````python
numbers = [10, 3, 7, 1, 9]
modify_elements(numbers)
print(numbers)
````

- The function receives a **reference** to the list `numbers`.
- Inside the loop, each element is multiplied by 2.
- The original list is updated in memory and no copy is made.


### Passing a Tuple to the Same Function
Tuples are **immutable**, so attempting to modify their contents causes a `TypeError`.


````python
numbers_tuple = (10, 20, 30)
print(numbers_tuple)

# Uncomment to see the error
# modify_elements(numbers_tuple)  #TypeError: 'tuple' object does not support item assignment
````

### Why the Error Happens
- Lists can be changed element by element (`items[i] = ...` is valid).  
- Tuples cannot, they don’t allow item assignment.  
- Trying to reassign a tuple element breaks immutability, leading to an error.


### Tuples with Mutable Elements
If a tuple **contains a mutable object** (like a list), the list inside the tuple *can* be modified.


````python
# Tuple containing a list
mixed_tuple = (10, [1, 2, 3])

# Modify the list inside the tuple
mixed_tuple[1][0] = 99
print(mixed_tuple)
````

### Understanding Tracebacks
When an error occurs, Python provides a **traceback** showing:
1. Where the error occurred (the exact line in the code).
2. The function call sequence that led to the error.
3. The specific line (`---->`) that caused the problem.

Example:
 
TypeError: 'tuple' object does not support item assignment

This tells us:
- The function `modify_elements()` tried to assign to a tuple element.
- The error happened on line 4 (the line with `items[i] *= 2`).
- Tuples can’t be modified — so the assignment failed.


## 5.8 Sorting Lists
**Sorting** arranges elements in a specific order typically **ascending** or **descending**.  
In Python, you can sort lists in-place using `.sort()` or create new sorted sequences using `sorted()`.


````python
# Example list
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]
````

### Sorting a List in Ascending Order
The list method `.sort()` arranges elements in **ascending order (smallest to largest)**.  
It modifies the original list directly.


````python
numbers.sort()
print(numbers)
````

### Sorting a List in Descending Order
To sort in **descending order**, pass the keyword argument `reverse=True`.


````python
numbers.sort(reverse=True)
print(numbers)
````

### Using the Built-In `sorted()` Function
`sorted()` returns a **new sorted list**, leaving the original sequence **unchanged**.  
It works with any iterable like - lists, tuples, or even strings.


````python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

ascending_numbers = sorted(numbers)
print("Sorted copy:", ascending_numbers)
print("Original list:", numbers)
````

### Sorting Strings
When sorting a string, Python orders the **characters alphabetically**.


````python
letters = 'fadgchjebi'
ascending_letters = sorted(letters)
print(ascending_letters)  
print(letters)             # original string unchanged
````

### Sorting Tuples
Tuples can also be sorted with `sorted()`.  
The result is always a **new list** (not a tuple).


````python
colors = ('red', 'orange', 'yellow', 'green', 'blue')
ascending_colors = sorted(colors)
print(ascending_colors)  
print(colors)             # original tuple remains the same
````

### Descending Order with `sorted()`
You can also sort in descending order using the same `reverse=True` keyword.


````python
descending_colors = sorted(colors, reverse=True)
print(descending_colors)  
````

## 5.9 Searching Sequences
Searching allows you to determine whether a sequence (like a list, tuple, or string)  
contains a value matching a given **key**.  
Python provides several built-in tools for searching efficiently.


### List Method `index()`
The `index()` method returns the **position (index)** of the first occurrence of a value in a list.  
If the value is not found, it raises a `ValueError`.


````python
numbers = [3, 7, 1, 4, 2, 8, 5, 6]

print(numbers.index(5))  
# print(numbers.index(100))  #  ValueError if not found
````

### Specifying a Starting Index
You can specify where the search begins by adding a **starting index** as a second argument.  
Here, we’ll duplicate the list first using `*=` to have more data to search.


````python
numbers *= 2
print(numbers)
# [3, 7, 1, 4, 2, 8, 5, 6, 3, 7, 1, 4, 2, 8, 5, 6]

# Search for 5 starting from index 7
print(numbers.index(5, 7))
````

### Specifying Starting and Ending Indices
You can also specify both **start** and **end** indices for a narrower search range.  
The search includes `start` but excludes `end`.


````python
# Search for 7 only within indices 0 to 3
# [3, 7, 1, 4, 2, 8, 5, 6, 3, 7, 1, 4, 2, 8, 5, 6]
print(numbers.index(7, 0, 4))
````

### Membership Operators: `in` and `not in`
Use `in` to check if a value exists in a sequence,  
and `not in` to verify that it does not.


````python
# [3, 7, 1, 4, 2, 8, 5, 6, 3, 7, 1, 4, 2, 8, 5, 6]
print(1000 in numbers)   
print(5 in numbers)      
print(1000 not in numbers)  
print(5 not in numbers)     
````

### Preventing `ValueError` When Searching
You can combine `in` with `index()` to search safely and avoid exceptions.


````python
key = 1000

if key in numbers:
    print(f'Found {key} at index {numbers.index(key)}')
else:
    print(f'{key} not found')  # Safe: avoids ValueError
````

### Built-In Functions `any()` and `all()`
These functions test **truth conditions** across iterable elements:

- `any(iterable)` → Returns `True` if *any* element is `True`.  
- `all(iterable)` → Returns `True` if *every* element is `True`.


````python
values = [0, 1, 2, 3]

print(any(values))  # True (because at least one nonzero value exists)
print(all(values))  # False (because 0 is considered False)
````

### Truthiness Recap
- Nonzero numbers → `True`
- Zero → `False`
- Non-empty collections (lists, tuples, strings) → `True`
- Empty collections → `False`


````python
print(any([]))          # False — empty iterable
print(all([1, 2, 3]))   # True — all nonzero
print(any([0, 0, 0]))   # False — all zero
````

## 5.10 Other List Methods
Python provides many list methods to add, remove, count, and manipulate list elements.  
Let’s explore the most useful ones with examples.


### Inserting an Element at a Specific Index
Use `.insert(index, value)` to insert an element at a specific position.  
Existing elements are shifted to the right.


````python
color_names = ['orange', 'yellow', 'green']
color_names.insert(0, 'red')  # Insert 'red' at the beginning
print(color_names)  
````

### Appending an Element to the End
Use `.append(value)` to add one element to the end of the list.


````python
color_names.append('blue')
print(color_names)  
````

### Adding Multiple Elements Using `extend()`
Use `.extend(iterable)` to add **all elements** of another iterable (like a list, string, or tuple) to the end.


````python
color_names.extend(['indigo', 'violet'])
print(color_names)
````

`extend()` is equivalent to using `+=` between lists.


### Extending with Strings and Tuples
Each character in a string or element in a tuple is added individually.


````python
sample_list = []
s = 'abc'
sample_list.extend(s)
print(sample_list)  # ['a', 'b', 'c']

t = (1, 2, 3)
sample_list.extend(t)
print(sample_list)  # ['a', 'b', 'c', 1, 2, 3]

# You can also pass a tuple directly
sample_list.extend((4, 5, 6))
print(sample_list)  # ['a', 'b', 'c', 1, 2, 3, 4, 5, 6]
````

### Removing the First Occurrence of a Value
Use `.remove(value)` to delete the **first occurrence** of a specific value.  
Raises a `ValueError` if the value isn’t found.


````python
color_names = ['orange', 'yellow', 'green']
color_names.remove('green')
print(color_names)
````

### Clearing a List
Use `.clear()` to remove all elements from a list.  
Equivalent to `color_names[:] = []`.


````python
color_names.clear()
print(color_names) 
````

### Counting Occurrences
Use `.count(value)` to count how many times a value appears in a list.


````python
responses = [1, 2, 5, 4, 3, 5, 2, 1, 3, 3, 1, 4, 3, 3, 3, 2, 3, 3, 2, 2]

for i in range(1, 6):
    print(f'{i} appears {responses.count(i)} times in responses')
````

### Reversing a List
Use `.reverse()` to reverse the order of list elements **in place** (modifies the original list).


````python
color_names = ['red', 'orange', 'yellow', 'green', 'blue']
color_names.reverse()
print(color_names) 
````

### Copying a List
Use `.copy()` to create a **shallow copy** of a list.  
This is equivalent to using slicing (`[:]`).


````python
copied_list = color_names.copy()
print(copied_list) 
````

### Summary of Common List Methods

| Method | Description | Example |
|:--|:--|:--|
| `insert(i, x)` | Inserts `x` at index `i` | `my_list.insert(2, 'a')` |
| `append(x)` | Adds `x` to end of list | `my_list.append(10)` |
| `extend(iterable)` | Adds all elements from another iterable | `my_list.extend([1,2,3])` |

| `count(x)` | Counts occurrences of `x` | `my_list.count(3)` |

## 5.11 Simulating Stacks with Lists
A **stack** is a data structure that stores elements in **Last-In, First-Out (LIFO)** order.  
The last item added (“pushed”) is the first item removed (“popped”).

Python doesn’t have a built-in stack type,  
but you can simulate one easily using a **list** and its built-in methods.


### Creating and Using a Stack
We’ll create an empty list called `stack`, then use:
- `.append(x)` → to **push** an item onto the stack  
- `.pop()` → to **pop** (remove and return) the most recent item


````python
stack = []  # create an empty stack

# Push elements onto the stack
stack.append('red')
print(stack)  # ['red']

stack.append('green')
print(stack)  # ['red', 'green']

# Pop elements (LIFO order)
print(stack.pop())  # 'green'
print(stack)        # ['red']

print(stack.pop())  # 'red'
print(stack)        # []
````

### Handling Empty Stack Errors
Attempting to pop from an empty list raises an `IndexError`,  
just like trying to access a nonexistent list index.


````python
# Uncomment to see the error
# stack.pop()  # IndexError: pop from empty list
````

To avoid this, check if the stack is not empty before popping:


````python
if len(stack) > 0:
    stack.pop()
else:
    print("Stack is empty — nothing to pop.")
````

### Summary
| Operation | Method | Description |
|:-----------|:--------|:-------------|
| **Push** | `append(x)` | Adds `x` to the top (end) of the stack |
| **Pop** | `pop()` | Removes and returns the top item |
| **Peek** | `stack[-1]` | Returns the top item *without* removing it |
| **Check empty** | `len(stack) == 0` | Tests if stack is empty |

---

Stacks are used in:
- **Function calls** (call stack)
- **Undo/Redo operations**
- **Expression parsing**
- **Backtracking algorithms**


## 5.12 List Comprehensions
**List comprehensions** are a compact, functional-style way to create new lists in a single line.  
They can replace many `for` loops that build lists using `.append()`.

The basic syntax is:

```python
[expression for item in iterable if condition]

This reads as:

“For each item in iterable, evaluate expression,
and include it in the new list if condition is true.”


### Creating a Simple List of Integers
Let’s start with a traditional `for` loop that appends numbers 1–5 into a list:


````python
list1 = []
for item in range(1, 6):
    list1.append(item)

print(list1)  # [1, 2, 3, 4, 5]
````

We can accomplish the same task in **one line** using a list comprehension:


````python
list2 = [item for item in range(1, 6)]
print(list2)  # [1, 2, 3, 4, 5]
````

This is equivalent to using `list(range(1, 6))`,  
but list comprehensions are more flexible when performing transformations.


### Mapping: Performing Calculations in a Comprehension
A comprehension can include an **expression** that transforms each element.

Here, we calculate the **cube** of each number from 1–5:


````python
list3 = [item ** 3 for item in range(1, 6)]
print(list3)  # [1, 8, 27, 64, 125]
````

This is an example of a **mapping operation**,  
where each element is *mapped* to a new value based on a function or formula.


### Filtering: Using `if` in a List Comprehension
You can add an **`if` condition** to include only elements that meet certain criteria.

Here, we create a list of **even numbers** from 1–10:


````python
list4 = [item for item in range(1, 11) if item % 2 == 0]
print(list4)  # [2, 4, 6, 8, 10]
````

The `if` clause filters the sequence,  
so only values that satisfy the condition (`item % 2 == 0`) are kept.


### Processing Another List
You can use list comprehensions to process *any iterable*, not just numbers.

Here, we create a list of colors in **uppercase**:


````python
colors = ['red', 'orange', 'yellow', 'green', 'blue']
colors2 = [color.upper() for color in colors]

print(colors2)  # ['RED', 'ORANGE', 'YELLOW', 'GREEN', 'BLUE']
print(colors)   # Original list remains unchanged
````

### Summary of List Comprehension Use Cases

| Operation | Description | Example |
|:-----------|:-------------|:-------------|
| **Basic creation** | Create list from range | `[x for x in range(5)]` |
| **Mapping** | Transform each item | `[x**2 for x in range(5)]` |
| **Filtering** | Include items meeting condition | `[x for x in range(10) if x % 2 == 0]` |
| **Processing iterables** | Apply function to each element | `[s.upper() for s in words]` |

---

**Benefits of list comprehensions:**
- Shorter and more readable than loops  
- Often faster than building lists with `append()`  
- Support both **mapping** (transforming) and **filtering** (selecting) operations  


## 5.13 Generator Expressions
**Generator expressions** look similar to list comprehensions but use **parentheses** `()` instead of square brackets `[]`.

They don’t create a full list in memory instead, they **generate values one at a time**, *on demand*.  
This behavior is called **lazy evaluation**, and it makes generators more **memory efficient** than list comprehensions.


### Greedy vs. Lazy Evaluation
- **List comprehensions** use **greedy evaluation** → they immediately build and store the entire list.  
- **Generator expressions** use **lazy evaluation** → they produce one value at a time as you iterate over them.

This is useful when working with **large datasets** or when you **don’t need all results at once**.


### Example: Generator Expression for Squaring Odd Numbers
Let’s square only the odd numbers in a list using a generator expression.


````python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

# Generator expression inside a for loop
# Iterate over a generator that yields the squares of odd numbers from 'numbers'
for value in (x ** 2 for x in numbers if x % 2 != 0):
    print(value, end=' ')
````

The generator produces one squared value at a time, instead of creating a full list of results.


### Storing a Generator Expression
If you assign a generator expression to a variable, it doesn’t create a list
it creates a **generator object**, which can be iterated over.


````python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

squares_of_odds = (x ** 2 for x in numbers if x % 2 != 0)
print(squares_of_odds)
````

This shows that `squares_of_odds` is a **generator object**, not a list.  
You can iterate over it using a loop or convert it into a list if needed.



````python
# Iterate through the generator
for num in squares_of_odds:
    print(num, end=' ')

# Or convert it to a list explicitly
numbers_list = list(x ** 2 for x in numbers if x % 2 != 0)
print("\nConverted list:", numbers_list)
````

### Summary: List Comprehension vs. Generator Expression

| Feature | List Comprehension | Generator Expression |
|:--|:--|:--|
| **Syntax** | `[expr for item in iterable if condition]` | `(expr for item in iterable if condition)` |
| **Evaluation Type** | Greedy (creates full list) | Lazy (produces values on demand) |
| **Memory Usage** | High (stores all values) | Low (one value at a time) |
| **Return Type** | `list` | `generator` |
| **Use Case** | When you need all results at once | When processing large data streams |

---

**Tip:**  
Use **list comprehensions** for small or reusable collections,  
and **generator expressions** for large or one-time sequences to save memory and improve performance.


## 5.14 Filter, Map and Reduce
Python provides several **functional-style programming tools** for working with sequences   
namely **`filter()`**, **`map()`**, and **`reduce()`**.

These allow you to process collections by **filtering**, **transforming**, and **reducing** data  
into simpler or summarized forms.


### Filtering Values with `filter()`
The built-in function **`filter(function, iterable)`** keeps only elements  
for which the given function returns `True`.

Let’s extract all **odd numbers** from a list.


````python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

# Define a helper function
def is_odd(x):
    """Returns True only if x is odd."""
    return x % 2 != 0

# Apply filter
odd_numbers = list(filter(is_odd, numbers))
print(odd_numbers) 
````

`filter()` is a **higher-order function** - it takes another function as input.  
It uses **lazy evaluation**, so you must wrap it in `list()` to view results.


### Using a Lambda Function
Instead of defining a named function, you can use a **lambda**  
a small, anonymous function written inline.


````python
# Equivalent result using a lambda expression
odd_numbers_lambda = list(filter(lambda x: x % 2 != 0, numbers))
print(odd_numbers_lambda) 
````

Example equivalence:

```python
def is_odd(x): return x % 2 != 0
# same as
lambda x: x % 2 != 0


### Mapping Values with `map()`
The built-in function **`map(function, iterable)`** transforms each element  
of an iterable using the given function.


````python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

# Square each number in the list
squares = list(map(lambda x: x ** 2, numbers))
print(squares)
````

`map()` also uses **lazy evaluation**, so wrap it in `list()` when you need all results.  
The equivalent list comprehension is:


````python
[item ** 2 for item in numbers]
````

### Combining `filter()` and `map()`
You can chain these functions together to **filter first**, then **map** the results.


````python
# Square only the odd numbers
odd_squares = list(map(lambda x: x ** 2, filter(lambda x: x % 2 != 0, numbers)))
print(odd_squares)
````

Step-by-step:
1. `filter(lambda x: x % 2 != 0, numbers)` → keeps only odd numbers.  
2. `map(lambda x: x ** 2, ...)` → squares those odd numbers.  
3. `list()` → materializes the results into a list.

The same can be written more clearly with a **list comprehension**:


````python
[x ** 2 for x in numbers if x % 2 != 0]
````

### Reducing Values with `sum()` and `reduce()`
A **reduction** processes a sequence into a **single value**,  
such as a total, minimum, or maximum.

Example using `sum()`:


````python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

total = sum(numbers)
print(total)  # 55
````

Python also provides a general-purpose **`reduce()`** function  
(from the `functools` module) for custom reductions.


````python
from functools import reduce

# Multiply all numbers together
# Use reduce() to multiply all elements in the list cumulatively: (((1*2)*3)*4)
product = reduce(lambda x, y: x * y, [1, 2, 3, 4])
print(product)  # 24
````

### Summary of Functional Tools

| Function | Purpose | Returns | Example |
|:--|:--|:--|:--|
| **filter(func, iterable)** | Keep only elements where `func(x)` is `True` | Iterator | `filter(lambda x: x>5, data)` |
| **map(func, iterable)** | Apply a function to every element | Iterator | `map(lambda x: x**2, data)` |
| **reduce(func, iterable)** | Combine all elements into one | Single value | `reduce(lambda a,b: a+b, data)` |

---

**Key Concepts**
- **filter →** select items  
- **map →** transform items  
- **reduce →** combine items  
- All use **lazy evaluation** — results are generated only when needed.  
- **List comprehensions** often combine these concepts in a cleaner, Pythonic way.


## 5.15 Other Sequence Processing Functions
Python includes several built-in functions that make it easier to process and compare sequence data.  
This section covers:
- Using `min` and `max` with a **key function**
- Iterating **backward** with `reversed()`
- Combining iterables with **`zip()`**


### Finding Minimum and Maximum with a Key Function
By default, `min()` and `max()` compare values directly, which can be misleading for **strings**  
because uppercase and lowercase letters have different **Unicode values**.

For example:


````python
'Red' < 'orange'  # True — but seems backward alphabetically!
````

````python
print(ord('R'))  # 82
print(ord('o'))  # 111
````

Since `'R'` (82) has a smaller Unicode value than `'o'` (111),  
Python considers `'Red'` < `'orange'`.

To compare strings **alphabetically**, we can use the `key` parameter  
to normalize case during comparison.


````python
colors = ['Red', 'orange', 'Yellow', 'green', 'Blue']

# Find alphabetic minimum and maximum
print(min(colors, key=lambda s: s.lower()))  # 'Blue'
print(max(colors, key=lambda s: s.lower()))  # 'Yellow'
````

The `key` argument specifies a one-parameter function used for comparison.  
Here, `lambda s: s.lower()` converts each string to lowercase for fair alphabetical comparison.


### Iterating Backward Through a Sequence
The **`reversed()`** function returns an iterator that traverses a sequence **in reverse** order.

Let’s use it to reverse and process a list.


````python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

# Create a new list of squares in reverse order
reversed_squares = [n ** 2 for n in reversed(numbers)]
print(reversed_squares)
````

`reversed()` does not modify the original list — it simply provides a backward iterator.


### Combining Iterables with `zip()`
The **`zip()`** function lets you iterate over multiple iterables **in parallel**.  
It returns tuples pairing elements with the same index from each iterable.


````python
names = ['Bob', 'Sue', 'Amanda']
grade_point_averages = [3.5, 4.0, 3.75]

# Iterate over both lists simultaneously
for name, gpa in zip(names, grade_point_averages):
    print(f'Name={name}; GPA={gpa}')
````

Each tuple contains elements from the same positions in each iterable.  
`zip()` stops when the **shortest iterable** runs out of elements.

You can also convert the zipped object to a list or tuple:


````python
paired = list(zip(names, grade_point_averages))
print(paired)
````

### Summary of Sequence Processing Functions

| Function | Description | Example |
|:--|:--|:--|
| **min(iterable, key=func)** | Returns smallest value (using optional key) | `min(colors, key=str.lower)` |
| **max(iterable, key=func)** | Returns largest value (using optional key) | `max(colors, key=str.lower)` |
| **reversed(iterable)** | Returns reversed iterator | `for x in reversed(numbers): ...` |
| **zip(iter1, iter2, ...)** | Combines multiple iterables into tuples | `zip(names, gpas)` |

---

**Key Takeaways**
- Use `key` with `min` and `max` to control comparison behavior (e.g., case-insensitive sorting).
- `reversed()` allows you to iterate backward efficiently.
- `zip()` is great for merging multiple sequences for parallel iteration.


## 5.16 Two-Dimensional Lists
A **two-dimensional list** (also called a *nested list* or *matrix*) is a list that contains **other lists** as its elements.  
It’s commonly used to represent **tables**, such as rows and columns of data.


### Creating a Two-Dimensional List
Let’s represent exam grades for 3 students who each took 4 exams.

Each **inner list** represents one student's grades (a row).


````python
a = [
    [77, 68, 86, 73],  # first student's grades
    [96, 87, 89, 81],  # second student's grades
    [70, 90, 86, 81]   # third student's grades
]
````

Here:
- `a[0]` → `[77, 68, 86, 73]` (first student’s grades)
- `a[1]` → `[96, 87, 89, 81]` (second student’s grades)
- `a[2]` → `[70, 90, 86, 81]` (third student’s grades)

Each element can be accessed with **two indices**:  
`a[row][column]`


````python
print(a[0][0])  
print(a[1][2])  
print(a[2][3])  
````

### Displaying a Two-Dimensional List
We can use **nested for loops** to print all elements row by row.


````python
for row in a:
    for item in row:
        print(item, end=' ')
    print()
````

### Understanding Nested Loops
To better understand how the loops work, we can print both the **indices** and **values**.


````python
for i, row in enumerate(a):         # outer loop for rows
    for j, item in enumerate(row):  # inner loop for columns
        print(f'a[{i}][{j}]={item}', end='  ')
    print()
````

**Explanation:**
- The outer loop iterates through each **row** (`i` = 0 → 2).  
- The inner loop iterates through each **column** (`j` = 0 → 3).  
- Together, they visit every cell in the 3×4 table.


### Two-Dimensional List Dimensions
A list with `m` rows and `n` columns is called an **m-by-n** list  
and has a total of `m × n` elements.

Example:


````python
rows = len(a)        # 3
columns = len(a[0])  # 4
print(f'This is a {rows}-by-{columns} list.')
````

### Summary
| Concept | Description | Example |
|:--|:--|:--|
| **2D List** | A list of lists | `[[1,2,3],[4,5,6]]` |
| **Access element** | Use two indices | `a[1][2]` |
| **Loop through rows** | `for row in a:` | iterates each row |
| **Loop through all values** | Nested loops with `enumerate` | prints row & column |
| **Size** | Rows × Columns | `len(a) × len(a[0])` |

---

**Tip:**  
While two-dimensional lists are useful for small tables,  
libraries like **NumPy** and **Pandas** offer faster and more convenient tools  
for working with large, structured, or numerical data.
